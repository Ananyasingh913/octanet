```python
import pickle

def create_record():
    print("\n")
    try:
        with open("atm.dat", 'ab') as f:
            while True:
                L = []
                uid = input("\nEnter the User ID: ")
                L.append(uid)
                pin = input("\nEnter the Pin: ")
                L.append(pin)
                balance = float(input("\nEnter the balance: "))
                L.append(balance)
                for _ in range(10):
                    L.append('Empty')
                pickle.dump(L, f)
                ch = input("Do you want to enter more records? (Y/N): ")
                if ch.lower() != 'y':
                    break
    except FileNotFoundError:
        print("File not found")

def id_check(user):
    flag = False
    try:
        with open("atm.dat", 'rb') as f:
            while True:
                try:
                    R = pickle.load(f)
                    if R[0] == user:
                        flag = True
                        break
                except EOFError:
                    break
    except FileNotFoundError:
        print("File not found")
    return flag

def pin_check(user, pin):
    try:
        with open("atm.dat", 'rb') as f:
            while True:
                try:
                    R = pickle.load(f)
                    if R[0] == user and R[1] == pin:
                        return True
                except EOFError:
                    break
    except FileNotFoundError:
        print("File not found")
    return False

def withdraw(user):
    try:
        with open("atm.dat", 'rb+') as f:
            A = float(input("Enter Amount to withdraw: "))
            while True:
                try:
                    pos = f.tell()
                    R = pickle.load(f)
                    if R[0] == user:
                        if R[2] >= A:
                            R[2] -= A
                            print("Money Withdrawn...Record updated")
                            for i in range(3, 13):
                                if R[i] == "Empty":
                                    R[i] = f"Withdrawn {A}"
                                    break
                            f.seek(pos)
                            pickle.dump(R, f)
                        else:
                            print("Insufficient Balance")
                        break
                except EOFError:
                    print("Record not found")
                    break
    except FileNotFoundError:
        print("File not found")

def deposit(user):
    try:
        with open("atm.dat", 'rb+') as f:
            A = float(input("Enter Amount to deposit: "))
            while True:
                try:
                    pos = f.tell()
                    R = pickle.load(f)
                    if R[0] == user:
                        R[2] += A
                        print("Money Deposited...Record updated")
                        for i in range(3, 13):
                            if R[i] == "Empty":
                                R[i] = f"Deposited {A}"
                                break
                        f.seek(pos)
                        pickle.dump(R, f)
                        break
                except EOFError:
                    print("Record not found")
                    break
    except FileNotFoundError:
        print("File not found")

def transaction_history(user):
    total_remaining = 0
    try:
        with open("atm.dat", "rb") as f:
            while True:
                try:
                    R = pickle.load(f)
                    if R[0] == user:
                        print("Transaction History for User:", user)
                        for i in range(3, 13):
                            if R[i] != "Empty":
                                print(R[i])
                        total_remaining = R[2]  # Get the current balance
                        break
                except EOFError:
                    print("Record not found")
                    break
    except FileNotFoundError:
        print("File not found")
        
    print("Total Remaining Balance:", total_remaining)


def transfer(user):
    try:
        with open("atm.dat", 'rb+') as f:
            A = float(input("Enter Amount to transfer: "))
            K = input("Enter Account Number to transfer: ")
            while True:
                try:
                    pos = f.tell()
                    R = pickle.load(f)
                    if R[0] == user:
                        if R[2] >= A:
                            R[2] -= A
                            print(f"Money Successfully transferred to account number {K}...Record updated")
                            for i in range(3, 13):
                                if R[i] == "Empty":
                                    R[i] = f"Transferred {A} to {K}"
                                    break
                            f.seek(pos)
                            pickle.dump(R, f)
                        else:
                            print("Insufficient Balance")
                        break
                except EOFError:
                    print("Record not found")
                    break
    except FileNotFoundError:
        print("File not found")

if __name__ == "__main__":
    create_record()
    chances = 3    
    user_id = input("Enter your User ID: ")
    if id_check(user_id):
        while chances != 0:
            user_pin = input("Enter your Pin: ")
            if pin_check(user_id, user_pin):
                while True:
                    print("\n\n\t\tMAIN MENU")
                    print("*")
                    print("1. Withdraw")
                    print("2. Deposit")
                    print("3. Transaction History")
                    print("4. Transfer")
                    print("5. Log Out")
                    print("*")

                    choice = int(input("Enter your choice between 1 to 5: "))

                    if choice == 1:
                        withdraw(user_id)
                    elif choice == 2:
                        deposit(user_id)
                    elif choice == 3:
                        transaction_history(user_id)
                    elif choice == 4:
                        transfer(user_id)
                    elif choice == 5:
                        print("\nSession Logged Out Successfully\nThank you")
                        break
                    else:
                        print("\nWrong Choice!\n\nTry again")
            else:
                chances -= 1
                print("Wrong Pin, Please Try Again...Attempts Left:", chances)
        else:
            print("\nSession Logged Out Successfully\nThank you")
    else:
        print("User ID not found. Please create an account.")
```
